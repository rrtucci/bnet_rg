# p0=0.7 Efficiency Plots for Ising-like dynamical Bnet

We plot the efficiency of our 2-dim Ising-like dynamical bnet for various temperatures and p0=0.7.

# p0 = 0.7

```python
beta_jj = BETA_JJ_CURIE * 20
jj = 1  # no jj dependence when h=0
beta = beta_jj / jj

p0 = 0.7
h=0
lam=0
for num_iter in [6,7]:
    net = Net(beta=beta, 
              jj=jj, 
              h=h, 
              lam=lam, 
              num_iter=num_iter, 
              p0=p0,
              do_reversing=False) #no change if do reversing
    net.do_plot("test.txt")
```

```python
# same as last one but smaller beta
beta_jj = BETA_JJ_CURIE * 3
jj = 1  # no jj dependence when h=0
beta = beta_jj / jj

for num_iter in [28,29]:
    net = Net(beta=beta, 
              jj=jj, 
              h=h, 
              lam=lam, 
              num_iter=num_iter, 
              p0=p0,
              do_reversing=False) #no change if do reversing
    net.do_plot("test.txt")
```
