```python
beta_jj = BETA_JJ_CURIE * 20
jj = 1  # no jj dependence when h=0
beta = beta_jj / jj
num_iter = 10

p0 = .3
h=0
lam=0
net = Net(beta=beta, 
          jj=jj, 
          h=h, 
          lam=lam, 
          num_iter=num_iter, 
          p0=p0,
          do_reversing=False) #no change if do reversing
net.do_plot("test.txt")
```
