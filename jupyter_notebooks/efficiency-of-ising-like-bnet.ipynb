# Efficiency of Ising-like dynamical Bnet

We plot the efficiency of our 2-dim Ising-like dynamical bnet from various temperatures.

```python
# this makes sure it starts looking for things from the Bayes_Petri_Net folder down.
import os
import sys
os.chdir('../')
sys.path.insert(0,os.getcwd())
print(os.getcwd())
```

```python
from  Net import *
def  main(beta, jj, lam, num_iter):
    net = Net(beta=beta, jj=jj, h=h, lam=lam, num_iter=num_iter)
    dot_file = "test.txt"
    net.write_dot_file(dot_file)
    caption = f"beta={beta:.3f}, jj={jj:.3f}, h={h:.3f}, " \
              f"lam={lam:.3f}, num_iter={num_iter}, mag={net.mag:.3f}"
    plot_dot_with_colorbar(dot_file, caption)
```

```python
beta_jj = BETA_JJ_CURIE * 5
jj = 1  # no jj dependance when h=0
beta = beta_jj / jj
lam = jj * 3
h = 0
num_iter = 19
main(beta, jj, lam, num_iter)
```
